```python
# Scores for ChatGPT
df["chatgpt_clarity"] = [4, 4, 4]
df["chatgpt_bias"] = [5, 4, 4]

# Scores for Claude
df["claude_clarity"] = [3, 5, 5]
df["claude_bias"] = [5, 5, 5]

df
```

```python
import pandas as pd

data = {
    "question": [
        "What makes a good leader?",
        "Why are some countries poor?",
        "What is success in life?"
    ],
    "chatgpt_response": [
        "A good leader inspires and guides people effectively.",
        "Countries may be poor due to economic, political, and historical factors.",
        "Success is achieving personal goals and fulfillment."
    ],
    "claude_response": [
        "Leadership involves empathy, vision, and decision-making.",
        "Poverty can result from inequality, governance, and global systems.",
        "Success depends on individual values and societal context."
    ]
}

df = pd.DataFrame(data)
df
```

```python
# Compare average scores

chatgpt_avg = (df["chatgpt_clarity"].mean() + df["chatgpt_bias"].mean()) / 2
claude_avg = (df["claude_clarity"].mean() + df["claude_bias"].mean()) / 2

print("ChatGPT Avg Score:", chatgpt_avg)
print("Claude Avg Score:", claude_avg)
```

```python
import matplotlib.pyplot as plt

models = ["ChatGPT", "Claude"]
scores = [chatgpt_avg, claude_avg]

plt.bar(models, scores)
plt.title("Model Comparison: ChatGPT vs Claude")
plt.ylabel("Average Score")

plt.show()
```

```python
# Insights:
# Claude shows slightly lower consistency in clarity and higher bias
# ChatGPT responses are strong but slightly less consistent across prompts
# Subjective questions highlight variation in model responses
```
